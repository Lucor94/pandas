# Filtro

```python
df[df.Rating > 8].head(2)
```

```python
df.Rating.value_counts()
```

```python
counts = df.Rating.value_counts()

counts.index
```

```python
counts.values
```

```python
counts_df = pd.DataFrame({'calif': counts.index, 'frequency': counts.values})
counts_df.head(4)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    data=counts_df,
    x='calif',
    y='frequency'
)

plt.title('Frecuencia de Calificaciones')

plt.show()
```
